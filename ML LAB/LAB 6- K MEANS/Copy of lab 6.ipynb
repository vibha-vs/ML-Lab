```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
from importlib import reload
plt=reload(plt)
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML LAB/LAB 6- K MEANS/income.csv')
df.head(10)
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['Age']])
df[['Age']] = scaler.transform(df[['Age']])

scaler.fit(df[['Income($)']])
df[['Income($)']] = scaler.transform(df[['Income($)']])
df.head(10)
```

```python
plt.scatter(df['Age'], df['Income($)'])
```

```python
k_range = range(1, 11)
sse = []
for k in k_range:
    kmc = KMeans(n_clusters=k)
    kmc.fit(df[['Age', 'Income($)']])
    sse.append(kmc.inertia_)
sse
```

```python
plt.xlabel = 'Number of Clusters'
plt.ylabel = 'Sum of Squared Errors'
plt.plot(k_range, sse)
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predict = km.fit_predict(df[['Age', 'Income($)']])
y_predict
```

```python
df['cluster'] = y_predict
df.head()
```

```python
df0 = df[df.cluster == 0]
df0
```

```python
df1 = df[df.cluster == 1]
df1
```

```python
df2 = df[df.cluster == 2]
df2
```

```python
km.cluster_centers_
```

```python
p1 = plt.scatter(df0['Age'], df0['Income($)'], marker='+', color='red')
p2 = plt.scatter(df1['Age'], df1['Income($)'], marker='*', color='blue')
p3 = plt.scatter(df2['Age'], df2['Income($)'], marker='^', color='green')
c = plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='black')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend((p1, p2, p3, c),
          ('Cluster 1', 'Cluster 2', 'Cluster 3', 'Centroid'))
```
